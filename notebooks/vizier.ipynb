## Vizier data

For a given target star, we can use vizier to check published data:
* Ca-II H/K=log(R'HK) or B-V mag color to use `BAFFLES`
* B-V mag color and log(R'HK) to use `EAGLES`
* Teff, Fe/H, logg + JHK photometry to use `isochrones`
* Teff and rotation period to use `gyro-interp`
* RA, Dec, parallax, proper motion, RV from Gaia to cross-match cluster membership in a given catalog

We can use the `Target` class to query vizier. It needs target coordinates in degrees.

```python
from astropy.coordinates import SkyCoord

coords = SkyCoord(toi['RA'], toi['Dec'], unit=["hourangle","deg"], frame='icrs')
coords
```

```python
coords.ra.deg, coords.dec.deg
```

```python
from kenmon import Target

t = Target(coords.ra.deg, coords.dec.deg)
t.query_vizier_param('log_EW8498_')
```
